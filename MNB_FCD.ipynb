```
nltk.download('stopwords')
from nltk.corpus import stopwords
indo = stopwords.words('indonesian')
```

```
corpus = []
for i in range(0, len(df)):

    # Penghilangan tanda baca (Tanda baca ditukar dengan spasi)
    review = re.sub('[^a-zA-Z]', ' ', df['Teks'][i])

    # Casefolding (semua huruf dalam case diubah menjadi lowercase)
    review = review.lower()
    review = review.split()

    # Penghilangan stopwords dalam Bahasa Indonesia
    review = [word for word in review if not word in indo]
    
    # Stemming (pengakaran kata-kata dalam corpora teks)
    review = [ps.stem(word) for word in review if not word in indo]

    review = ' '.join(review)
    corpus.append(review)
```

```
for kata in corpus:
    print(kata)
    print()
```

```
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import binarize
cv = CountVectorizer()
X = cv.fit_transform(corpus).toarray()
col = cv.get_feature_names()
new_frame = pd.DataFrame(X, columns = col)
new_frame
```

```
D = df.shape[0]
tabel_fcd = []
# fcd_test = np.zeros(len(col))
df_total = np.zeros(len(col)) 
df0 = np.zeros(len(col)) 
df1 = np.zeros(len(col))
df2 = np.zeros(len(col))
df3 = np.zeros(len(col))
df4 = np.zeros(len(col))

for t_ctr in range(0,len(col)):
  for d_ctr in range(0,D):
    df_total[t_ctr] += X[d_ctr,t_ctr]
    if (df['Topik'][d_ctr] == 0):
      df0[t_ctr] += X[d_ctr,t_ctr]
    if (df['Topik'][d_ctr] == 1):
      df1[t_ctr] += X[d_ctr,t_ctr]
    if (df['Topik'][d_ctr] == 2):
      df2[t_ctr] += X[d_ctr,t_ctr]
    if (df['Topik'][d_ctr] == 3):
      df3[t_ctr] += X[d_ctr,t_ctr]
    if (df['Topik'][d_ctr] == 4):
      df4[t_ctr] += X[d_ctr,t_ctr]

fcd0 = np.zeros(len(col)) 
fcd1 = np.zeros(len(col))
fcd2 = np.zeros(len(col))
fcd3 = np.zeros(len(col))
fcd4 = np.zeros(len(col))

for fcd_tctr in range(0,len(col)):
  fcd0[fcd_tctr] = (2*df0[fcd_tctr] - df_total[fcd_tctr]) / df_total[fcd_tctr]
  fcd1[fcd_tctr] = (2*df1[fcd_tctr] - df_total[fcd_tctr]) / df_total[fcd_tctr]
  fcd2[fcd_tctr] = (2*df2[fcd_tctr] - df_total[fcd_tctr]) / df_total[fcd_tctr]
  fcd3[fcd_tctr] = (2*df3[fcd_tctr] - df_total[fcd_tctr]) / df_total[fcd_tctr]
  fcd4[fcd_tctr] = (2*df4[fcd_tctr] - df_total[fcd_tctr]) / df_total[fcd_tctr]

tabel_fcd.append(fcd0)
tabel_fcd.append(fcd1)
tabel_fcd.append(fcd2)
tabel_fcd.append(fcd3)
tabel_fcd.append(fcd4)
```
